```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].value_counts()
```

```python
df.shape
```

```python
x = df.drop(['species'],axis=1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=765243)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression

model_Dt = DecisionTreeClassifier()
model_lr = LogisticRegression()

model_Dt.fit(x_train,y_train)
model_lr.fit(x_train,y_train)
```

```python
y_pred_dt = model_Dt.predict(x_test)
y_pred_dt
```

```python
y_pred_lr = model_lr.predict(x_test)
y_pred_lr
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred_dt),"is the accuracy of Decision trees")
print(accuracy_score(y_test,y_pred_lr),"is the accuracy of Logistic Regression")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_dt))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_lr))
```

```python
from sklearn import tree

fig = plt.figure(figsize=(15,12))
tree.plot_tree(model_Dt, feature_names=df.drop('species',axis=1).columns, class_names=df['species'].unique().tolist(), fontsize=10, filled=True)
plt.show()
```
